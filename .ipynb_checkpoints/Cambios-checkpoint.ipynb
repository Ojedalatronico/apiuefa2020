```python
from pymongo import MongoClient
from dotenv import load_dotenv
import pandas as pd
import os
import json

load_dotenv()
user=os.getenv("MONGO_USER")
password=os.getenv("MONGO_PASS")

URL=f"mongodb+srv://{user}:{password}@uefa.kgpdy.mongodb.net"

db=MongoClient(URL).get_database("Project")
```

```python
UEFA=db["uefa2020_results"].find({},{"_id":0})
data=pd.DataFrame(UEFA)
data.head()
```

```python
event=data["events_list"]
```

```python
event=pd.Series(event)
s=event[0]
s
```

```python
s = s.replace("[", "")
s = s.replace("]", "")
s = s.replace("'", "\"")
a=json.loads(s)
```

```python
Goal=0
Yellow_card=0
Red_card=0
Substitution=0
for i in event:
    e=i.split(",")
    for lineas in e:
        if "Goal" in lineas:
            Goal+=1
        if "Yellow card" in lineas:
            Yellow_card+=1
        if "Red card" in lineas:
            Red_card+=1
        if "Substitution" in lineas:
            Substitution+=1
print(Goal)
print(Yellow_card)
print(Red_card)
print(Substitution)

"""for i in s:
    if i == "{":
        contador=1
    if contador==1:
        linea=cadena+str(i)
    if i == "}":
        contador=0
        print(linea)
        break"""
```
