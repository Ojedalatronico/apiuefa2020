```python
import pandas as pd
import ast
df = pd.read_csv("euros2020_results_cleaned.csv")
```

```python
df.head()
```

```python
season="2020"
```

```python
pens=df["pens"].value_counts("True")*51
total_pens=pens[1]
```

```python
total_match=df["date"].count()
```

```python
total_shots=df["total_shots_home"].sum()+df["total_shots_away"].sum()
average_shots=total_shots/total_match
```

```python
total_shots_on_target=df["shots_on_target_home"].sum()+df["shots_on_target_away"].sum()
average_shots_on_target=total_shots_on_target/total_match
```

```python
total_goals=df["team_home_score"].sum()+df["team_away_score"].sum()
average_goals=total_goals/total_match
```

```python
total_yellow_cards=df["yellow_cards_away"].sum()+df["yellow_cards_home"].sum()
average_yellow_cards=total_yellow_cards/total_match
```

```python
total_red_cards=df["red_cards_away"].sum()+df["red_cards_home"].sum()
average_red_card=total_red_cards/total_match
```

```python
total_cards=total_yellow_cards+total_red_cards
average_cards=total_cards/total_match
```

```python
data=[season,total_match,total_shots,average_shots,total_shots_on_target,average_shots_on_target,total_goals,average_goals,total_yellow_cards,average_yellow_cards,total_red_cards,average_red_card,total_cards,average_cards]
columnas=["season","total_match","total_shots","average_shots","total_shots_on_target","average_shots_on_targe","total_goals","average_goals","total_yellow_cards","average_yellow_cards","total_red_cards","average_red_card","total_cards","average_cards"]
df=pd.Series(data)
```

```python
df.index=[columnas]
```

```python
df=df.to_frame().T
```
